## .str 함수

### split

```python
df['diet'].str.split(' ') # 리스트 형태 
```

```python
df['diet'].str.split(' ', expand=True) # 각각 컬럼으로 분리
```

```python
df['diet'].str.contains('low') # 대소문자 주의
```
